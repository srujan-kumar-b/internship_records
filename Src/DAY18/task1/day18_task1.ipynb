```python
import sqlite3
import pandas as pd
```

```python
conn = sqlite3.connect("internship.db")
cursor = conn.cursor()
```

```python
cursor.execute("""
CREATE TABLE interns1 (
    intern_id INTEGER PRIMARY KEY,
    intern_name TEXT,
    track TEXT,
    stipend INTEGER
)
""")
```

```python
cursor.executemany("""
INSERT INTO interns1 (intern_id, intern_name, track, stipend)
VALUES (?, ?, ?, ?)
""", [
    (1, "jeevan", "AI", 7000),
    (2, "rakshith", "Web", 4000),
    (3, "shamanth", "AI", 8000),
    (4, "anusha", "Data Science", 6000),
    (5, "raj", "Data Science", 4500)
])

conn.commit()
```

```python
query1 = """
SELECT *
FROM interns1
WHERE track = 'Data Science'
AND stipend > 5000
"""

df1 = pd.read_sql_query(query1, conn)
df1
```

```python
query2 = """
SELECT track, AVG(stipend) AS avg_stipend
FROM interns1
GROUP BY track
"""

df2 = pd.read_sql_query(query2, conn)
df2
```

```python
query3 = """
SELECT track, COUNT(*) AS total_interns
FROM interns1
GROUP BY track
"""

df3 = pd.read_sql_query(query3, conn)
df3
```
